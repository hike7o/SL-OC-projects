__FONCTION DEFINED TO HELP DESCRIBING THE CHARACTERISTICS OF THE INPUT FILES__

```python
def duplicates(data):
    """Calculate the number of duplicates in dataframe data"""

    return len(data)-len(data.drop_duplicates())


def missing_cells(data):
    """Calculate the number of empty cells in dataframe data"""

    return data.isna().sum().sum()


def missing_cells_percent(data):
    """Calculate the percentage of empty cells in dataframe data"""

    return missing_cells(data)/(data.size)*100


def filled_cells(data):
    """Calculate the number of existing values in dataframe data"""

    return data.notna().sum()


def filled_cells_percent(data):
    """Calculate the percentage of existing values in dataframe data"""

    return filled_cells(data)/(len(data.index))*100
```

La segmentation RFM ou mthode RFM est une mthode de segmentation principalement dveloppe  l'origine pour les actions de marketing direct des vadistes et qui s'applique dsormais galement aux acteurs du e-commerce et du commerce traditionnel.

La segmentation RFM prend en compte la Rcence (date de la dernire commande), la Frquence des commandes et le Montant (de la dernire commande ou sur une priode donne) pour tablir des segments de clients homognes.

La segmentation RFM permet de cibler les offres, d'tablir des segments bass sur la valeur des clients et de prvenir l'attrition en identifiant des segments  risque.

Les donnes RFM restent des donnes trs utilises, mais une "simple" segmentation RFM est de plus en plus rare. Les donnes RFM sont dsormais intgres dans des procdures de ciblage et de segmentation comportant de plus en plus de variables et de donnes. Elles dpassent dsormais le cadre du marketing direct et peuvent s'appliquer  la publicit display en utilisant une DMP et des procdures de CRM onboarding.


## 2. DATAFRAMES CONNECTION

### 1. ORDER TO ORDER items on order_id

```python
item_order = pd.merge(orders, order_items, how="left", on='order_id')
item_order
```

### 2. PAYMENTS TO ORDERS on order_id

```python
item_order_pay = pd.merge(item_order, order_payments,
                          on='order_id')
item_order_pay
```

### 3. REVIEWS TO ORDERS/PAYMENTS on order_id

```python
pay_review = pd.merge(order_reviews, item_order_pay, on='order_id')
pay_review
```

### 4. CUSTOMERS on order_id

```python
pay_review_cust = pd.merge(customers, pay_review,
                           on='customer_id')
pay_review_cust
```

### 5. SELLERS on seller_id

```python
sell_cust = pd.merge(pay_review_cust, sellers,
                     on='seller_id')
sell_cust
```

### 6. PRODUCTS on product_id

```python
sell_cust_products = pd.merge(sell_cust, products,
                              on='product_id')
sell_cust_products
```

### 7. TRANSLATION on product_id

```python
ord_items_products_eng = pd.merge(sell_cust_products, translation,
                                  on='product_category_name')
ord_items_products_eng
```

```python
# We drop the product category name in portuguese to work with the english one
df_gross = ord_items_products_eng.drop('product_category_name',
                                       axis=1)
df_gross.info()
```
